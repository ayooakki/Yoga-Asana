```python

# Remove stopwords and lemmatize others
for i in range (len(sentences)):
  words_in_this=nltk.word_tokenize(sentences[i])
  words_in_this=[lemmatizer.lemmatize(word) for word in words_in_this if word not in set(stopwords.words('english')) ]
  for word in words_in_this:
    words_lemmatized.append(word)
  sentences_lemmatized[i]=' '.join(words_in_this)

print(sentences_lemmatized)  
print(words_lemmatized)
# words and sentences updated wrt lemmatization
```

```python
# clean the paragraph
import re

corpus=[] #stores the final cleaned list of sentences

for i in range(len(sentences)):
  this_sentence=re.sub('[^a-zA-Z]',' ',sentences[i]) # substitute everything other than alphabets with spaces
  
  this_sentence=this_sentence.lower() # convert to lower case

  words_in_this_sen=this_sentence.split() # split into list of words

  words_in_this_sen=[lemmatizer.lemmatize(word) for word in words_in_this_sen if word not in set(stopwords.words('english')) ] #lemmatize
  
  this_sentence=' '.join(words_in_this_sen)
  corpus.append(this_sentence)
  
df[x1]=(corpus)  
```
